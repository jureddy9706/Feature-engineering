# Count Or Frequency Encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
column=[1,3,5,6,7,8,9,13]
```

```python
train=train_set[column]
train
```

```python
train.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train.head()
```

```python
len(train['Employment'].unique())
```

```python
for feature in train.columns:
    print(feature,':',len(train[feature].unique()))
```

```python
 train.head()
```

```python
train['Country'].value_counts()
```

```python
train['Country'].replace(train['Country'].value_counts(),inplace=True) # replaced
```

```python
train.head()
```

# Target guided Ordinal Encoding

In this technique, we will transform our categorical variable by comparing it to the target or output variable.

Steps:

1) Choose a categorical variable.

2) Take the aggregated mean of the categorical variable and apply it to the target variable.

3) Assign higher integer values or a higher rank to the category with the highest mean.
